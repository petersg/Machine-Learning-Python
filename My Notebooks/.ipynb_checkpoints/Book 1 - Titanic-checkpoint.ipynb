# Cargar datos desde funcion open


```python
data3 = open(mainpath +"/" + "customer-churn-model/Customer Churn Model.txt", "r")
```

```python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1
    
print ("El dataSet tiene %d filas y %d columnas" %(counter,n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

# Lectura y escritura de ficheros

```python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Columns.txt"
```

Separando las comas por tabulaciones o /t y uniendolas

```python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.split(",")
            outfile1.write("\t".join(fields))
```

# Cargamos los datos con GET desde la web

```python
def downloadFromURL(url, fiename):
    import urllib3

    http = urllib3.PoolManager()
    r = http.request("GET", medals_url)
    r.status
    respuesta = r.data
    #Convierto los datos en string
    str_dato=respuesta.decode("utf-8")

    #Divido string en array de filas
    filas = str_dato.split("\n")

    #Extraigo la 1era linea que sera la cabecera
    colIndice = filas[0].split(",")
    numCols = len(colIndice)
    print("Hay un total de %d " %numCols + "lineas")

    #Generamos Diccionario vacio
    contadorDic= 0
    dicPrinc = {}
    for col in colIndice:
        dicPrinc[col] = []

    #Procesar fila a fila para ir rellenando la informacion en el diccionario
    for fila in filas:
        if(contadorDic > 0):
            #Dividimos string por comas
            valor = fila.strip().split(",")
            #aÃ±adimos el valor a la columna
            for i in range(len(colIndice)):
                dicPrinc[colIndice[i]].append(valor[i])
        contadorDic+=1

    print("Nuestro data set tiene %d filas y %d columnas" %(contadorDic,numCols))
    
    return medals_data 
```

```python
df_medallas = pd.DataFrame(dicPrinc)
df_medallas.head()
```
